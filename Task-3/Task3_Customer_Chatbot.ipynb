```python
def chatbot_response(user_input):
    user_input = user_input.lower()

    if "hours" in user_input or "working" in user_input:
        return "Our working hours are 9 AM to 6 PM, Monday to Friday."

    elif "contact" in user_input or "support" in user_input:
        return "You can contact support at support@example.com."

    elif "services" in user_input:
        return "We provide customer support, sales assistance, and technical help."

    elif "hello" in user_input or "hi" in user_input:
        return "Hello! How can I help you today?"

    else:
        return "Sorry, I didn’t understand that. Please contact support."
```

Chatbot Response Logic :

This cell defines the chatbot’s response function .
User input is converted to lowercase for consistent keyword matching.
The chatbot identifies user intent using keyword-based conditions (hours, contact, services, greetings).
Based on the detected intent, an appropriate customer support response is returned.
A fallback response is included for unknown queries to ensure smooth interaction.

```python
import streamlit as st

st.title("Customer Support Chatbot")

user_input = st.text_input("Ask your question:")

if user_input:
    response = chatbot_response(user_input)
    st.write("Chatbot:", response)
```
